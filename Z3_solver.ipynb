1. Кот, пес и мышь - пример головоломки


```python
# наши переменные (количества животных)
dog, cat, mouse = Ints('dog cat mouse')
solve(dog >= 1,   # не менее одной собаки
      cat >= 1,   # не менее одного кота
      mouse >= 1, # не менее одной мыши

      dog + cat + mouse == 100, # всего ровно сто животных

      # У нас ровно 100 долларов (10000 центов):
      #   пёс стоит 15 долларов (1500 центов),
      #   кот стоит 1 доллар (100 центов),
      #   мышь стоит 25 центов
      1500 * dog + 100 * cat + 25 * mouse == 10000)
```

3. Задача про автомобили

Владелец автомагазина продал два автомобиля, первый с прибылью в 10%, а второй с убытком в 10%. В результате этой сделки владелец автомагазина получил 5% прибыли. Во сколько обошёлся каждый из автомобилей владельцу магазина, если прибыль составила 1000 долларов?

```python
car1, car2 = Reals('стоимость_1_машины стоимость_2_машины')

print("Общая прибыль: 1000 долларов")

solve(
    1.1 * car1 + 0.9 * car2 == 1.05 * (car1 + car2),
    0.05 * (car1 + car2) == 1000
)
```

4. Двадцать конфет

Томми, Уилли, Мэгги и Энн купили на 20 центов 20 конфет. Шоколадные конфеты стоят 4 цента штука, леденцы продаются по цене 4 штуки за цент, а шоколадные драже стоят 1 цент пара.
Сколько конфет каждого сорта купили дети?



```python
lolipop, chocolate, dragees = Reals('леденцы шоколадные шоколадные_дражже')

solve(
    lolipop + chocolate + (2 * dragees) == 20, # Всего 20 конфет
    lolipop == 20 - chocolate - 2 * dragees,
    4 * chocolate + ((20 - chocolate - 2 * dragees) / 4) + dragees == 20
)
```

Задача по физике.

Привычный маршрут торгового катера — выехать со своей пристани, проплыть 77 км до соседней пристани против течения реки и вернуться обратно. Обратный путь занимает на 4 ч меньше. Чему равна скорость течения, если собственная скорость катера составляет 9 км/ч?


```python
x = Real('Vтеч')
# S = V*T
# 77 / 9 + x - скорость по течению
# 77 / 9 - x - скорость против течению

solve(
    77 / 9 - x == 77 / 9 + x - 4,
    x != 9,
    x != -9,
)
```

5. Задача по физике

Запрос:
Привычный маршрут торгового катера — выехать со своей пристани, проплыть 77 км до соседней пристани против течения реки и вернуться обратно. Обратный путь занимает на 4 ч меньше. Чему равна скорость течения, если собственная скорость катера составляет 9 км/ч?

### Ответ

Решение

Пусть скорость течения равна v км/ч.

Тогда
• скорость катера против течения: 9 − v
• по течению: 9 + v

Время в пути:
• туда: 77 ÷ (9 − v)
• обратно: 77 ÷ (9 + v)

По условию разница во времени — 4 часа:

77 ÷ (9 − v) − 77 ÷ (9 + v) = 4

Решаем уравнение:

1. Приведём к общему знаменателю:
77 × (18 ÷ (81 − v²)) = 4
2. Сократим 18 и 4:
77 × (9 ÷ (81 − v²)) = 2
3. Умножаем:
693 ÷ (81 − v²) = 2
4. Перемножаем крест‑накрест:
693 = 2 × (81 − v²)
5. Раскрываем:
693 = 162 − 2v²
6. Переносим:
2v² = 162 − 693 = −531
Здесь видно, что выражение получилось отрицательным. Значит, где‑то ошибка.
Проверяем вычисления внимательно:

Правильно будет так:

1. 77 ÷ (9 − v) − 77 ÷ (9 + v) = 4
2. Выносим 77:
77 × ( (9 + v − (9 − v)) ÷ ( (9 − v)(9 + v) ) ) = 4
3. В числителе остаётся 2v:
77 × (2v ÷ (81 − v²)) = 4
4. Сократим 2:
77v ÷ (81 − v²) = 2
5. Перемножаем:
77v = 2 × (81 − v²)
6. Раскрываем:
77v = 162 − 2v²
7. Переносим всё в одну сторону:
2v² + 77v − 162 = 0
Решаем квадратное уравнение:

• дискриминант:
77² + 4 × 2 × 162 = 5929 + 1296 = 7225
• корень из дискриминанта: 85

Теперь корни:

v = (−77 + 85) ÷ (4) = 8 ÷ 4 = 2
второй корень отрицательный, не подходит

Ответ
Скорость течения равна 2 км/ч.
